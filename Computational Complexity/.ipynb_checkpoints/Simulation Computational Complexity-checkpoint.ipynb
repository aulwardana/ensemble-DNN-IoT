```python
import matplotlib.pyplot as plt
import numpy as np

# Average values from the provided data
T_DNN = 190  # Average training time for one DNN model (in seconds)
P_DNN = 100.44  # Average prediction time for one model (in seconds)
NumDevices = 9  # Average number of IoT devices
len_models = 9   # Average total number of models

# Varying number of IoT devices with decimal values
num_devices = np.arange(1, 20, 0.5)

# With retaining
training_complexity_with_retaining = num_devices * NumDevices * T_DNN
prediction_complexity_with_retaining = num_devices * len_models * P_DNN
averaging_complexity_with_retaining = num_devices * len_models
total_complexity_with_retaining = (
    training_complexity_with_retaining +
    prediction_complexity_with_retaining +
    averaging_complexity_with_retaining
)

# Without retaining (constant number of devices)
num_devices_without_retaining = 9  # Keeping the number of devices constant
num_devices_without_retaining_array = np.full_like(num_devices, num_devices_without_retaining)

# Set training complexity to a constant value
training_complexity_without_retaining = np.full_like(num_devices, NumDevices * T_DNN)
prediction_complexity_without_retaining = np.full_like(num_devices, len_models * P_DNN)
averaging_complexity_without_retaining = (
    num_devices_without_retaining_array * len_models
)
total_complexity_without_retaining = (
    training_complexity_without_retaining +
    prediction_complexity_without_retaining +
    averaging_complexity_without_retaining
)

# Plotting
plt.figure(figsize=(12, 8))

# Plot with retaining
plt.subplot(2, 1, 1)
plt.plot(num_devices, training_complexity_with_retaining, label='Training (With Retaining)')
plt.plot(num_devices, prediction_complexity_with_retaining, label='Prediction (With Retaining)')
plt.plot(num_devices, averaging_complexity_with_retaining, label='Averaging (With Retaining)')
plt.plot(num_devices, total_complexity_with_retaining, label='Total (With Retaining)')

plt.title('With Retaining')
plt.xlabel('Number of IoT Devices')
plt.ylabel('Computational Complexity (seconds)')
plt.legend()
plt.grid(True)

# Plot without retaining
plt.subplot(2, 1, 2)
plt.plot(
    num_devices,
    training_complexity_without_retaining,
    label='Training (Without Retaining)',
    linestyle='--'
)
plt.plot(
    num_devices,
    prediction_complexity_without_retaining,
    label='Prediction (Without Retaining)',
    linestyle='--'
)
plt.plot(
    num_devices,
    averaging_complexity_without_retaining,
    label='Averaging (Without Retaining)',
    linestyle='--'
)
plt.plot(
    num_devices,
    total_complexity_without_retaining,
    label='Total (Without Retaining)',
    linestyle='--'
)

plt.title('Without Retaining')
plt.xlabel('Number of IoT Devices')
plt.ylabel('Computational Complexity (seconds)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
